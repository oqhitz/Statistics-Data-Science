Hypothesis Testing : Do setosa species have larger **sepal width** than versicolor ?

```python
setosa = df[df['Species'] == 'Iris-setosa'][['Species', 'SepalWidthCm']]
versicolor = df[df['Species'] == 'Iris-versicolor'][['Species', 'SepalWidthCm']]
print('Iris Setosa Species Sepal Width data: ')
print(setosa)
print('\nIris Versicolor Species Sepal Width data: ')
print(versicolor)
```

First thing first, let's check for outliers

```python
plt.figure(figsize=(5,3))
sns.boxplot(setosa['SepalWidthCm'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(versicolor['SepalWidthCm'])
plt.show()
```

since we don't see any outlier in those boxplots, we can use mean / average as central tendency of the data

```python
setosa['SepalWidthCm'].mean()
```

```python
versicolor['SepalWidthCm'].mean()
```

before we go to the hypothesis test, let's check distribution of both datas with density plot

```python
plt.figure(figsize=(5,3))
sns.distplot(setosa['SepalWidthCm'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(setosa['SepalWidthCm'])
plt.show()
```

from both density plot of setosa and versicolor species, we can see that both datas distribution is close to normal

Hypothesis Testing :

Does setosa species has larger sepal width than versicolor species ?

Null Hypothesis / H0 : **setosa sepal width <= versicolor sepal width**

Alternative Hypothesis / H1 : **setosa sepal width > versicolor sepal width**

Both variables are independent, so we will perform hypothesis test with independent samples t-Test

```python
setosa_SW = setosa['SepalWidthCm']
versicolor_SW = versicolor['SepalWidthCm']
```

```python
stats, pvalue = st.ttest_ind(setosa_SW, versicolor_SW)
```

```python
alpha = 0.05
print('P-Value = ', pvalue)
if pvalue >= alpha:
    print('Insufficient to reject H0')
else:
    print('Sufficient to reject H0')
```

***
Since we failed to reject H0,
the result of the test is : ***its proven that setosa sepal width is larger than versicolor's***
